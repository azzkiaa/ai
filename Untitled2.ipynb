## Menghitung Nilai RFM

### Subtask:
Membangun fitur Recency, Frequency, dan Monetary untuk setiap pelanggan. Recency adalah jumlah hari sejak pembelian terakhir, Frequency adalah total jumlah transaksi, dan Monetary adalah total nilai uang yang dihabiskan.


**Reasoning**:
First, I need to ensure the 'InvoiceDate' column is of datetime type to perform date calculations. Based on the last kernel state, it appears to be an object type, so I will convert it.



````python
import pandas as pd

# Convert 'InvoiceDate' to datetime if it's not already
# Use errors='coerce' to turn unparseable dates into NaT
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')

# Drop rows where InvoiceDate conversion failed (if any)
df.dropna(subset=['InvoiceDate'], inplace=True);

# Determine the snapshot date as one day after the last InvoiceDate
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1);

# Group by 'Customer ID' to calculate RFM values
rfm_df = df.groupby('Customer ID').agg(
    Recency=('InvoiceDate', lambda date: (snapshot_date - date.max()).days),
    Frequency=('Invoice', 'nunique'),
    Monetary=('TotalPrice', 'sum')
).reset_index()

# Display the first few rows of the RFM DataFrame and its information
display(rfm_df.head())
display(rfm_df.info())
````

## Menangani Nilai Negatif pada Kolom Quantity

### Subtask:
Memfilter transaksi dengan nilai 'Quantity' negatif, karena ini biasanya mengindikasikan retur barang dan tidak relevan untuk analisis pembelian.


**Reasoning**:
To address the subtask of handling negative 'Quantity' values, I will filter the DataFrame `df` to keep only rows where the 'Quantity' is greater than 0, as negative quantities typically represent returns and are not relevant for purchase analysis.



````python
df = df[df['Quantity'] > 0]

# Display the first few rows of the filtered DataFrame and its information
display(df.head())
df.info()
````

## Normalisasi Fitur RFM

### Subtask:
Melakukan normalisasi pada fitur RFM agar setiap fitur memiliki skala yang sama, penting untuk PCA dan K-Means.


**Reasoning**:
I need to normalize the 'Recency', 'Frequency', and 'Monetary' columns in the `rfm_df` DataFrame using `StandardScaler` to ensure all features have a similar scale, which is crucial for PCA and K-Means. This involves importing the scaler, applying it to the selected columns, and then creating a new DataFrame with the scaled values.



````python
from sklearn.preprocessing import StandardScaler

# Select the RFM features for scaling
rfm_features = rfm_df[['Recency', 'Frequency', 'Monetary']]

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply StandardScaler to the RFM features
scaled_features = scaler.fit_transform(rfm_features)

# Create a new DataFrame with the scaled features
rfm_scaled_df = pd.DataFrame(scaled_features, columns=['Recency_scaled', 'Frequency_scaled', 'Monetary_scaled'])

# Display the first few rows of the scaled DataFrame
display(rfm_scaled_df.head())
````

## Menerapkan PCA (Principal Component Analysis)

### Subtask:
Mereduksi dimensi fitur RFM yang telah dinormalisasi menggunakan PCA untuk memudahkan visualisasi dan mempercepat proses clustering.


**Reasoning**:
I need to import the PCA module from `sklearn.decomposition` to perform Principal Component Analysis on the scaled RFM features.



````python
from sklearn.decomposition import PCA

# Initialize PCA with 2 components for visualization purposes
pca = PCA(n_components=2)

# Apply PCA to the scaled RFM features
pca_features = pca.fit_transform(rfm_scaled_df)

# Create a new DataFrame with the PCA components
rfm_pca_df = pd.DataFrame(data=pca_features, columns=['PC1', 'PC2'])

# Display the explained variance ratio for each component
print("Explained variance ratio per component:", pca.explained_variance_ratio_)
print("Cumulative explained variance:", pca.explained_variance_ratio_.sum())

# Display the first few rows of the PCA-transformed DataFrame
display(rfm_pca_df.head())
````

## Menentukan Jumlah Klaster Optimal

### Subtask:
Menggunakan metode Elbow atau Silhouette Score untuk menentukan jumlah klaster (K) yang paling optimal untuk algoritma K-Means.


**Reasoning**:
To determine the optimal number of clusters using the Silhouette Score, I will first import the necessary modules, then iterate through a range of cluster numbers (k), fit a KMeans model for each k, calculate the silhouette score, and store it for visualization.



````python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Create an empty list to store silhouette scores
silhouette_scores = []

# Define a range of possible cluster numbers (e.g., from 2 to 10)
k_range = range(2, 11)

# Loop through the range of k values
for k in k_range:
    # Initialize KMeans model
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Added n_init to suppress warning

    # Fit the KMeans model to the rfm_pca_df DataFrame
    kmeans.fit(rfm_pca_df)

    # Predict the cluster labels
    cluster_labels = kmeans.predict(rfm_pca_df)

    # Calculate the silhouette score
    score = silhouette_score(rfm_pca_df, cluster_labels)

    # Append the score to the list
    silhouette_scores.append(score)

# Plot the silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal K')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_range)
plt.grid(True)
plt.show()

print("Silhouette Scores:", silhouette_scores)
````

## Menerapkan K-Means Clustering

### Subtask:
Melakukan K-Means clustering pada data yang telah direduksi dimensinya untuk mengelompokkan pelanggan.


**Reasoning**:
I need to perform K-Means clustering using the optimal number of clusters determined from the Silhouette Score, which is 2. Then, I will add the resulting cluster labels to both the PCA-transformed DataFrame and the original RFM DataFrame.



````python
from sklearn.cluster import KMeans

# Initialize KMeans model with n_clusters=2 (optimal from Silhouette Score)
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)

# Fit the KMeans model to the rfm_pca_df DataFrame
kmeans.fit(rfm_pca_df)

# Add the cluster labels to rfm_pca_df as a new column named 'Cluster'
rfm_pca_df['Cluster'] = kmeans.labels_

# Add the same cluster labels to rfm_df
rfm_df['Cluster'] = kmeans.labels_

# Display the first few rows of rfm_pca_df with clusters
display(rfm_pca_df.head())

# Display the first few rows of rfm_df with clusters
display(rfm_df.head())
````

````python
# Step 2 & 3: Clean and Convert 'Quantity' and 'Price'
# First, ensure Quantity and Price are numeric. Use errors='coerce' to handle bad values.
df['Quantity'] = df['Quantity'].astype(str).str.replace(',', '.').apply(pd.to_numeric, errors='coerce')
df['Price'] = df['Price'].astype(str).str.replace(',', '.').apply(pd.to_numeric, errors='coerce')

# Remove rows where Quantity or Price became NaN due to conversion errors
df.dropna(subset=['Quantity', 'Price'], inplace=True)

# Filter out rows where Quantity is less than or equal to 0
df = df[df['Quantity'] > 0]

# Step 4: Recalculate 'TotalPrice'
df['TotalPrice'] = df['Quantity'] * df['Price']

# Step 5: Clean and Convert 'InvoiceDate'
# Use dayfirst=True and errors='coerce' to handle mixed formats and unparseable dates
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], dayfirst=True, errors='coerce')

# Remove rows where InvoiceDate became NaT
df.dropna(subset=['InvoiceDate'], inplace=True)

# Step 6: Handle Missing 'Customer ID'
df.dropna(subset=['Customer ID'], inplace=True)

# Ensure 'Customer ID' is a consistent type, e.g., integer, for grouping
df['Customer ID'] = df['Customer ID'].astype(str).str.replace(',', '').apply(pd.to_numeric, errors='coerce')
df.dropna(subset=['Customer ID'], inplace=True)
df['Customer ID'] = df['Customer ID'].astype(int)

# Display info after comprehensive cleaning
print("DataFrame Info after Cleaning:")
df.info()

# Display head after cleaning
print("\nDataFrame Head after Cleaning:")
display(df.head())

# Step 7: Recalculate RFM values
# Determine the snapshot date as one day after the last InvoiceDate
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

# Group by 'Customer ID' to calculate RFM values
rfm_df = df.groupby('Customer ID').agg(
    Recency=('InvoiceDate', lambda date: (snapshot_date - date.max()).days),
    Frequency=('Invoice', 'nunique'),
    Monetary=('TotalPrice', 'sum') # Monetary will now be float due to TotalPrice being float
).reset_index()

# Display the first few rows of the RFM DataFrame and its information
print("\nRFM DataFrame Info:")
display(rfm_df.head())
rfm_df.info()

# Step 8: Normalize RFM Features
from sklearn.preprocessing import StandardScaler

# Select the RFM features for scaling
rfm_features = rfm_df[['Recency', 'Frequency', 'Monetary']]

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply StandardScaler to the RFM features
scaled_features = scaler.fit_transform(rfm_features)

# Create a new DataFrame with the scaled features
rfm_scaled_df = pd.DataFrame(scaled_features, columns=['Recency_scaled', 'Frequency_scaled', 'Monetary_scaled'])

# Display the first few rows of the scaled DataFrame
print("\nScaled RFM Features:")
display(rfm_scaled_df.head())

# Step 9: Apply PCA
from sklearn.decomposition import PCA

# Initialize PCA with 2 components for visualization purposes
pca = PCA(n_components=2)

# Apply PCA to the scaled RFM features
pca_features = pca.fit_transform(rfm_scaled_df)

# Create a new DataFrame with the PCA components
rfm_pca_df = pd.DataFrame(data=pca_features, columns=['PC1', 'PC2'])

# Display the explained variance ratio for each component
print("\nExplained variance ratio per component:", pca.explained_variance_ratio_)
print("Cumulative explained variance:", pca.explained_variance_ratio_.sum())

# Display the first few rows of the PCA-transformed DataFrame
print("\nPCA-Transformed RFM Features:")
display(rfm_pca_df.head())

# Step 10: Determine Optimal Clusters (using Silhouette Score as per plan)
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Create an empty list to store silhouette scores
silhouette_scores = []

# Define a range of possible cluster numbers (e.g., from 2 to 10)
k_range = range(2, 11)

# Loop through the range of k values
for k in k_range:
    # Initialize KMeans model
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)

    # Fit the KMeans model to the rfm_pca_df DataFrame
    kmeans.fit(rfm_pca_df)

    # Predict the cluster labels
    cluster_labels = kmeans.predict(rfm_pca_df)

    # Calculate the silhouette score
    score = silhouette_score(rfm_pca_df, cluster_labels)

    # Append the score to the list
    silhouette_scores.append(score)

# Plot the silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal K')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_range)
plt.grid(True)
plt.show()

print("\nSilhouette Scores for K values:", silhouette_scores)
# Determine optimal k based on the highest silhouette score
optimal_k = k_range[silhouette_scores.index(max(silhouette_scores))]
print(f"Optimal number of clusters (K) based on Silhouette Score: {optimal_k}")

# Step 11: Apply K-Means Clustering with optimal K
# Initialize KMeans model with optimal_k
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)

# Fit the KMeans model to the rfm_pca_df DataFrame
kmeans.fit(rfm_pca_df)

# Add the cluster labels to rfm_pca_df as a new column named 'Cluster'
rfm_pca_df['Cluster'] = kmeans.labels_

# Add the same cluster labels to rfm_df
rfm_df['Cluster'] = kmeans.labels_

# Display the first few rows of rfm_pca_df with clusters
print("\nPCA-Transformed RFM with Clusters:")
display(rfm_pca_df.head())

# Display the first few rows of rfm_df with clusters
print("\nRFM DataFrame with Clusters:")
display(rfm_df.head())

# Step 12: Analyze and Visualize Clusters

# Analyze characteristics of each cluster (average RFM values)
cluster_analysis = rfm_df.groupby('Cluster')[['Recency', 'Frequency', 'Monetary']].mean()
print("\nAverage RFM Values per Cluster:")
display(cluster_analysis)

# Visualisasi segmen pelanggan di ruang PCA
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=rfm_pca_df, palette='viridis', s=100, alpha=0.7)
plt.title('Customer Segments (PCA-Reduced)', fontsize=16)
plt.xlabel('Principal Component 1 (PC1)', fontsize=12)
plt.ylabel('Principal Component 2 (PC2)', fontsize=12)
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

# Box Plot untuk Recency, Frequency, Monetary per Klaster
plt.figure(figsize=(18, 6))

# Box Plot for Recency per Cluster
plt.subplot(1, 3, 1)
sns.boxplot(x='Cluster', y='Recency', data=rfm_df, palette='viridis', hue='Cluster', legend=False)
plt.title('Recency by Cluster', fontsize=14)
plt.xlabel('Cluster', fontsize=12)
plt.ylabel('Recency (Days)', fontsize=12)

# Box Plot for Frequency per Cluster
plt.subplot(1, 3, 2)
sns.boxplot(x='Cluster', y='Frequency', data=rfm_df, palette='viridis', hue='Cluster', legend=False)
plt.title('Frequency by Cluster', fontsize=14)
plt.xlabel('Cluster', fontsize=12)
plt.ylabel('Frequency (Number of Invoices)', fontsize=12)

# Box Plot for Monetary per Cluster
plt.subplot(1, 3, 3)
sns.boxplot(x='Cluster', y='Monetary', data=rfm_df, palette='viridis', hue='Cluster', legend=False)
plt.title('Monetary by Cluster', fontsize=14)
plt.xlabel('Cluster', fontsize=12)
plt.ylabel('Monetary (Total Spend)', fontsize=12)

plt.tight_layout()
plt.show()

# Distribusi Pelanggan di Setiap Klaster
cluster_counts = rfm_df['Cluster'].value_counts().sort_index()
plt.figure(figsize=(8, 6))
sns.barplot(x=cluster_counts.index, y=cluster_counts.values, palette='viridis', hue=cluster_counts.index, legend=False)
plt.title('Distribution of Customers in Each Cluster', fontsize=16)
plt.xlabel('Cluster', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print("\nDistribution of Customers in Each Cluster:")
display(cluster_counts)

# Final Task: Summarize insights
print("\n## Final Task: Ringkasan Hasil Analisis Klaster Pelanggan dan Pola Pembelian Mereka")
print("### Analisis Klaster Pelanggan RFM")
print("Berdasarkan analisis RFM (Recency, Frequency, Monetary) dan K-Means clustering pada data yang telah direduksi dimensinya menggunakan PCA, pelanggan berhasil disegmentasikan menjadi dua klaster optimal.")
print("\n#### Karakteristik Klaster:")
print(cluster_analysis.to_markdown(numalign="left", stralign="left"))

print("\n#### Interpretasi Klaster:")
print("1.  **Cluster 0 (Pelanggan Tidak Aktif/Berisiko):** Klaster ini memiliki Recency rata-rata tertinggi, menunjukkan bahwa pelanggan di segmen ini melakukan pembelian paling baru dalam jangka waktu yang lebih lama. Frequency dan Monetary rata-rata mereka juga tergolong rendah. Pelanggan di klaster ini cenderung kurang aktif atau berisiko churn.")
print("2.  **Cluster 1 (Pelanggan Aktif/Nilai Tinggi):** Klaster ini memiliki Recency rata-rata terendah, menunjukkan pembelian yang sangat baru. Frequency dan Monetary rata-rata mereka sangat tinggi, mengindikasikan bahwa ini adalah pelanggan yang sangat sering berbelanja dan menghabiskan banyak uang. Mereka adalah pelanggan inti yang loyal dan berharga.")
print("\n#### Implikasi Bisnis dan Rekomendasi:")
print("Segmentasi ini memberikan wawasan penting untuk strategi pemasaran yang ditargetkan:")
print("-   **Untuk Klaster 0:** Fokus pada strategi re-engagement seperti penawaran khusus untuk pembelian berikutnya, personalisasi rekomendasi produk berdasarkan riwayat pembelian terbatas, atau survei untuk memahami alasan ketidakaktifan.")
print("-   **Untuk Klaster 1:** Terapkan program loyalitas eksklusif, penawaran premium, atau layanan pelanggan khusus untuk mempertahankan kepuasan dan mendorong pembelian berulang. Pelanggan ini adalah pendorong utama pendapatan dan harus diberi prioritas.")
print("\nVisualisasi menunjukkan pemisahan yang jelas antara kedua klaster di ruang PCA dan perbedaan signifikan dalam nilai-nilai RFM mereka, menegaskan keberhasilan segmentasi ini.")
````

## Hitung Nilai RFM

### Subtask:
Bangun fitur Recency, Frequency, dan Monetary untuk setiap pelanggan berdasarkan data yang telah dibersihkan dan dikonversi.


**Reasoning**:
To specifically address the subtask of calculating RFM values, I will regenerate the RFM DataFrame (`rfm_df`) using the already cleaned and converted `df`. This involves determining a snapshot date, grouping by 'Customer ID', and then aggregating 'InvoiceDate' for Recency, unique 'Invoice' for Frequency, and 'TotalPrice' for Monetary. Finally, I will display the head and info of the resulting `rfm_df` to verify the calculations.



````python
import pandas as pd

# 1. Tentukan snapshot_date sebagai satu hari setelah tanggal InvoiceDate terakhir
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

# 2. Kelompokkan df berdasarkan kolom 'Customer ID'
# 3. Hitung 'Recency', 'Frequency', dan 'Monetary'
rfm_df = df.groupby('Customer ID').agg(
    Recency=('InvoiceDate', lambda date: (snapshot_date - date.max()).days),
    Frequency=('Invoice', 'nunique'),
    Monetary=('TotalPrice', 'sum')
).reset_index()

# 7. Tampilkan lima baris pertama dari DataFrame RFM yang baru dibuat (rfm_df) dan informasi umum
print("\nRFM DataFrame Head:")
display(rfm_df.head())
print("\nRFM DataFrame Info:")
rfm_df.info()
````

## Menerapkan PCA (Principal Component Analysis)

### Subtask:
Reduksi dimensi fitur RFM yang telah dinormalisasi menggunakan PCA menjadi dua komponen utama.


**Reasoning**:
To reduce the dimensionality of the normalized RFM features, I will apply PCA with two components as instructed. This involves importing PCA, initializing it, fitting and transforming the scaled data, and then displaying the results along with the explained variance.

## Menentukan Jumlah Klaster Optimal

### Subtask:
Gunakan metode Silhouette Score untuk menentukan jumlah klaster (K) yang paling optimal untuk algoritma K-Means.


**Reasoning**:
To determine the optimal number of clusters, I will use the Silhouette Score method. This involves iterating through a range of possible cluster numbers, fitting a K-Means model for each, calculating its silhouette score, and then visualizing these scores to identify the optimal K.



````python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Create an empty list to store silhouette scores
silhouette_scores = []

# Define a range of possible cluster numbers (e.g., from 2 to 10)
k_range = range(2, 11)

# Loop through the range of k values
for k in k_range:
    # Initialize KMeans model
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Added n_init to suppress warning

    # Fit the KMeans model to the rfm_pca_df DataFrame
    kmeans.fit(rfm_pca_df)

    # Predict the cluster labels
    cluster_labels = kmeans.predict(rfm_pca_df)

    # Calculate the silhouette score
    score = silhouette_score(rfm_pca_df, cluster_labels)

    # Append the score to the list
    silhouette_scores.append(score)

# Plot the silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal K')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_range)
plt.grid(True)
plt.show()

print("Silhouette Scores:", silhouette_scores)
# Determine optimal k based on the highest silhouette score
optimal_k = k_range[silhouette_scores.index(max(silhouette_scores))]
print(f"Optimal number of clusters (K) based on Silhouette Score: {optimal_k}")
````

## Menerapkan K-Means Clustering

### Subtask:
Lakukan K-Means clustering pada data yang telah direduksi dimensinya menggunakan jumlah klaster optimal yang ditemukan.


**Reasoning**:
To perform K-Means clustering as instructed, I will initialize the KMeans model with the optimal number of clusters, fit it to the PCA-reduced data, and then assign the resulting cluster labels to both the PCA DataFrame and the original RFM DataFrame for further analysis.



````python
from sklearn.cluster import KMeans

# Initialize KMeans model with n_clusters=optimal_k (from Silhouette Score), random_state=42, n_init=10
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)

# Fit the KMeans model to the rfm_pca_df DataFrame
kmeans.fit(rfm_pca_df)

# Add the cluster labels to rfm_pca_df as a new column named 'Cluster'
rfm_pca_df['Cluster'] = kmeans.labels_

# Add the same cluster labels to rfm_df
rfm_df['Cluster'] = kmeans.labels_

# Display the first few rows of rfm_pca_df with clusters
print("\nPCA-Transformed RFM with Clusters:")
display(rfm_pca_df.head())

# Display the first few rows of rfm_df with clusters
print("\nRFM DataFrame with Clusters:")
display(rfm_df.head())
````
